```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('Crop_recommendation.csv')
datahead=data.head(10)
datahead
```

```python
data_label=data["label"]
data_label
```

```python
from sklearn import preprocessing
```

```python
print(data['label'].head(20))
label_encoder=preprocessing.LabelEncoder()
data['label']=label_encoder.fit_transform(data['label'])
print("after label data")
print(data['label'].head(20))
```

```python
x=data.drop('label',axis=1) #input
y=data['label'] #output(dependent data)
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2,test_size=0.2)
#random_state=2 (which means retesting the values again)
```

```python
print("total data size:",data.shape[0])
print(x_train.shape[0])
print(x_test.shape[0])
```

```python
from sklearn import linear_model
from sklearn import metrics
#pip install scikit-learn (command to use in vs code)
```

```python
logreg=linear_model.LogisticRegression(solver='lbfgs',C=1.5)
logistic=logreg.fit(x_train,y_train)
y_pred=logistic.predict(x_train)
print("accuracy:",metrics.accuracy_score(y_train,y_pred))
```

```python
CM_lr=metrics.confusion_matrix(y_train,y_pred)
CM_lr
```

```python
TP=CM_lr[0][0]
FP=CM_lr[0][0]
TN=CM_lr[0][0]
FN=CM_lr[0][0]
print("TP:",TP)
print("FP:",FP)
print("TN:",TN)
print("FN:",FN)
print("total:",(TP+FP+TN+FN))
# in a multi-class scenario, TP, FP, TN, and FN are not as straightforward as in a binary classification.(gemini)
accuracy = CM_lr.diagonal().sum() / CM_lr.sum()
# Calculate overall accuracy from the confusion matrix
print("Accuracy:", accuracy)
#accuracy calculation acc to sir=((TP+TN+FN)/(TP+FP+TN+FN))*100
```

```python
prec=metrics.precision_score(y_train,y_pred,average='micro')
print("precision:",prec)
#prec=TP/(TP+FP)*100
#sensitivity=(TP/(TN+FP))*100
```
